```python
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt
```

## 연습 문제 제안
문제 1: 특정 날짜의 데이터 가져오기
python
코드 복사
### 질문: 2023년 5월 15일의 데이터를 가져오세요.

```python
# TikTok 주가 데이터 다운로드
ticker = "TTOO"  # TikTok (티커 코드 예시)
tiktok_data = yf.download(ticker, start="2020-01-01", end="2023-12-31")
tiktok_data.columns = tiktok_data.columns.droplevel(1) #tricker 열 삭제
# 데이터 확인
print("TikTok 주가 데이터 샘플:\n", tiktok_data.head())
print(tiktok_data.info())
```

```python
specific_date_data = tiktok_data.loc['2022-01-03']
print("\n2022-01-03의 데이터:\n", specific_date_data)
specific_date_data = tiktok_data.loc['2023-05-15']
print("\n2023-05-15의 데이터:\n", specific_date_data)
```

문제 2: 특정 기간의 데이터 가져오기
python
코드 복사
### 질문: 2022년 전체 데이터만 추출하세요.

```python
# 문제 2: 특정 기간(1년)의 데이터 가져오기
# one_year_data = tiktok_data.loc['2022-01-01':'2022-12-31']
one_year_data = tiktok_data.loc['2022']
print("\n2021년 TikTok 데이터:\n", one_year_data.head())
```

문제 3: 조건에 맞는 데이터 필터링
python
코드 복사
### 질문: 종가가 5000달러 이상인 날의 데이터를 추출하세요.

```python
# 문제 3: 종가(Close)가 5000달러 이상인 날의 데이터 필터링
high_price_data = tiktok_data[tiktok_data.Close >= 5000]
print("\n종가가 5000달러 이상인 데이터:\n", high_price_data.head())
high_price_data.info()
```

문제 4: 거래량이 많은 날 분석
python
코드 복사
### 질문: 거래량이 1백만 이상인 데이터를 추출하고, 해당 데이터의 평균 종가를 계산하세요.

```python
# 문제 4: 거래량(Volume)이 1,000,000 이상인 날의 데이터 필터링
high_volume_data = tiktok_data[tiktok_data.Volume >= 1000000]
avg1 = high_volume_data['Close'].mean()
print("\n거래량이 1,000,000 이상인 데이터:\n", high_volume_data.head())
print(avg1)
```

```python
high_volume_data.loc[:, ['Close', 'Open']].plot()
```

문제 5: 시각화
python
코드 복사
### 질문: 2021년 데이터를 사용하여 종가(Close)와 거래량(Volume)의 트렌드를 시각화하세요.

```python
plt.figure(figsize=(12, 6))
plt.plot(one_year_data.index, one_year_data["Close"], label="Close Price", color="blue")
plt.bar(one_year_data.index, one_year_data["Volume"], label="Volume", color="orange", alpha=0.5)
plt.title("TikTok Stock Close Price and Volume in 2021")
plt.xlabel("Date")
plt.ylabel("Value")
plt.legend()
plt.show()

plt.figure(figsize=(12, 6))
plt.plot(one_year_data.index, one_year_data["Close"], label="TikTok Close Price (2021)", color="purple")
plt.title("TikTok Close Prices in 2021")
plt.xlabel("Date")
plt.ylabel("Close Price")
plt.legend()
plt.grid()
plt.show()
```
