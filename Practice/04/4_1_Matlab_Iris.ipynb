```python
import pandas as pd
import seaborn as sns
iris = sns.load_dataset("iris")
iris.head()
```

### 데이터셋의 구조
iris 데이터셋은 4개의 특징(특성)과 1개의 타겟(클래스)으로 구성됩니다.

- 특징(Features)
  - 꽃받침 길이(sepal length) (단위: cm)
  - 꽃받침 폭(sepal width) (단위: cm)
  - 꽃잎 길이(petal length) (단위: cm)
  - 꽃잎 폭(petal width) (단위: cm)

- 타겟(Target)
  + 붓꽃의 품종 (총 3가지 클래스):
    - Setosa (세토사)
    - Versicolor (버시칼라)
    - Virginica (버지니카)

```python
df = iris.copy()
df['target'] = pd.Categorical(iris.species).codes
df
```

```python
iris.columns = ['sepal_length(cm)', 'sepal_width(cm)', 'petal_length(cm)', 'petal_width(cm)', 'class_name']
iris.head()
```

```python
def target(c):
  result = []
  for s in c:
    if s == 'setosa':
      result.append(0)
    elif s == 'versicolor':
      result.append(1)
    else:
      result.append(2)
  return result

iris['target'] = target(iris['class_name'])
iris = iris[['sepal_length(cm)', 'sepal_width(cm)', 'petal_length(cm)', 'petal_width(cm)', 'target', 'class_name']]
iris
```
